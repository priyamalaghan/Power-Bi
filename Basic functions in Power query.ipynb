## Text Functions

### Split and Trim_Text
- Get the data
- Click transform, Power Query window is opened
- Select Column
- In Transform menu bar, click 'Split column' option and Click on 'By Delimiter'
- Select or enter the delimiter and click 'ok'
- To Change the text to upper/lower/Capitalize case, trim(To remove unwanted space) or clean the data, to add Prefix/Suffix to data click format option 
- To extract left,right, mid text
    - Click extract option
    - Choose required option
- Click 'close and apply' button
- Click on 'Refresh' button on any changes in original dataset

- Using Add Column menu bar, changes are applied by adding new column
- From Applied steps box, Click on setting of required step to customize the column formula
- view menu bar -> formula bar
- To count rows
    - Transform menu bar -> Count rows

## Date Functions

- Get the data
- Click transform/Add column, Power Query window is opened
- Select the Columns
- Date functions: Year, Quarter, Month, week, day
- FewExamples
    - Example1: To find year,qtr,month,day of date format mm/dd/yyyy
        - Click on Date option and select required option to extract day,week,month,qtr,year
        - Note: When you select multiple date columns, few available options are Substract Days, Earliest/Latest(on comparing two date columns)
    - To get Name of day and month of date
        - Date -> Day -> Name of Day,
        - Date -> Month -> Name of Month
    - To get Day of Week/Month/Year and Week of Month/Year
        - Date -> Day -> Day of Week/Day of Year
        - Date -> Week -> Week of year/Week of month
    - To extract Date from Date and Time
        - Date -> Date Only
        - Time -> Time Only
    - To get age from date
        - Date -> Age , Select column(Age in number of days), Duration -> Total Years
    - Which day of year, Quarter, Month your Date of Birth is
        - Date -> Day -> Day : day of month
        - Date -> Year -> Start of year , Date of birth - Start of year = Day of year
        - Date -> Quarter -> Start of Quarter , Date of birth - Start of Quarter = Day of Quarter

## Appending Files and Tables in Power-Bi

- **Append multiple CSV Files from a folder**
    - Copy the URL/Path where multiple CSV files are stored
    - Click on Get data and then click 'more'
    - Click All and then select 'Folder'and click connect
    - Paste the path
    - Now click on required options combine & transform data/combine & Load, Load, Transform data
    - Click on refresh preview in Home tab, if any new file is added to folder
    - Close and apply

- **Append multiple excel tables/sheets from single file**
    - Get Data : Excel Sheet with multiple tables
    - Here you can have different number of columns
    - Select all sheets while loading the data to PBI desktop
    - Home tab -> Append queries -> Append queries as new/Append queries
    - Click on refresh preview in Home tab, if any new table/row/Column is added
    - Close and apply

- **Append Excel tables with different number of columns**
    - The position of columns might not be same
    - Here you can have different number of columns
    - Click on get data and open excel
    - Prefer to select table instead of sheets
    - Click on transform data
    - Home tab -> Append queries -> Append queries as new/Append queries
    - The column position doesnot matter, the column name should be same
    - Even if you have different number of columns, the number of appending steps would remain same
    - Close and apply

- **Append multiple Excel files from a folder**
    - The position of columns might not be same
    - Here you can have different number of columns
    - Copy the URL/Path where multiple CSV files are stored
    - Click on Get data and then click 'more'
    - Paste the path
    - Click on transform data, from this meta data only Content and Name columns are important
    - Click Add Column menu bar -> Custom Column -> Write M-func i.e = Excel.Workbook([Content]), (remeber this is case sensitive) -> Click ok, a new column named 'Custom' is created -> Click on double headed arrow -> Click ok
    - Select table from 'Custom.Kind' column dropdown
    - Select 'Name','Custom.Data' and 'Custom.Item' columns and remove other columns
    - Click on double headed arrow of 'Custom.Data' column -> uncheck ' Use original column name as prefix' ->  Click ok
    - Close and apply
    - Click on 'Refresh' button on any new additional file in folder
    - Close and apply

- **Append different data source files in power query**
    - Let us consider we have two data source Excel and CSV
    - Get Data -> Excel -> open -> Load
    - Get Data -> CSV -> open -> Load
    - Home tab -> Transform Data -> Append queries as new
    - Select tables and click ok
    - Close and apply

## Merging Files and Tables in Power-Bi

- **Merge Tables/Sheets in power query**
    - Get Data -> Excel
    - Select the required tables
    - Click on Transform Data
    - Home menu bar -> Merge Queries -> Merge Queries as new 
    - Select the relationship between both the tables, choose join kind, click ok
    - New query is created
    - Click double headed arrow of'MasterTable' column -> From dropdown select required columns

- **Merge Data from multiple Excel Files/Workbooks in Power Query**
    - Let us consider below example
    - Get Data -> Excel
    - Load two tables from two different excel files
    - Click transform table menu bar
    - Home menu bar -> Merge Queries -> Merge Queries as new 
    - Select the relationship between both the tables, choose join kind, click ok
    - New query is created
    - Click double headed arrow of'MasterTable' column -> From dropdown select required columns

- **Merge Data from different Data Sources in Power Query**
    - Let us consider below example
    - Get Data -> Excel
    - Click on transform data instead of load data
    - Now you are in power query
    - Click new source -> click text/csv -> click ok
    - Home menu bar -> Merge Queries -> Merge Queries as new 
    - Select the relationship between both the tables, choose join kind, click ok
    - New query is created
    - Click double headed arrow of'MasterTable' column -> From dropdown select required columns

- **Merge data having multiple matching columns or multiple criteria in Power Query**
    - Let us consider below example
    - Get Data -> Excel
    - Click on transform data instead of load data
    - Now you are in power query
    - Click new source -> click excel -> click ok
    - Home menu bar -> Merge Queries -> Merge Queries as new 
    - Select the relationship between both the tables using multiple matching columns, choose join kind, click ok
        - Note: sequence of selecting the multiple matching column from both tables must be same
    - New query is created
    - Click double headed arrow of'MasterTable' column -> From dropdown select required columns
